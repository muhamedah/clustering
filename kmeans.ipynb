```python
# Data manipulation and numerical operations
import pandas as pd
import numpy as np

# Visualization libraries
import matplotlib.pyplot as plt  # Static plots
import seaborn as sns  # Enhanced statistical visualizations

# Clustering and distance calculations
from scipy.cluster.hierarchy import dendrogram, linkage  # Hierarchical clustering
from scipy.spatial.distance import pdist  # Pairwise distances

# Machine learning algorithms and preprocessing
from sklearn.cluster import KMeans  # K-means clustering
from sklearn.preprocessing import StandardScaler  # Feature scaling
from sklearn.decomposition import PCA  # Dimensionality reduction
```

<h3>display using scatter plot<h3>

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))  # 1 row, 3 columns, smaller figure


# ---- Flame ----
axes[0].scatter(df_flame.iloc[:, 0], df_flame.iloc[:, 1])
axes[0].set_title("Flame Dataset")
axes[0].set_xlabel("Feature 1")
axes[0].set_ylabel("Feature 2")

# ---- Blobs ----
axes[1].scatter(df_blobs.iloc[:, 0], df_blobs.iloc[:, 1])
axes[1].set_title("Blobs Dataset")
axes[1].set_xlabel("Feature 1")
axes[1].set_ylabel("Feature 2")

# ---- Varied ----
axes[2].scatter(df_varied.iloc[:, 0], df_varied.iloc[:, 1])
axes[2].set_title("Varied Dataset")
axes[2].set_xlabel("Feature 1")
axes[2].set_ylabel("Feature 2")

plt.tight_layout()
plt.show()
```
